```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('ipl 2020/final_players_ipl_2020_with_info.csv')
df2 = pd.read_csv('ipl 2021/final_players_ipl_2021_with_info.csv')
df3 = pd.read_csv('ipl 2022/final_players_ipl_2022_with_info.csv')
df4 = pd.read_csv('ipl 2023/final_players_ipl_2023_with_info.csv')
df5 = pd.read_csv('ipl 2024/final_players_ipl_2024_with_info.csv')
df6 = pd.read_csv('t20 2022/final_players_details_t20_2024.csv')
df7 = pd.read_csv('t20 2024/final_players_details_t20_2024.csv')
df8 = pd.read_csv('t20 2020/final_players_details_t20_2020.csv')
df9 = pd.read_csv('ODI 2019/final_players_ODI_2019_info.csv')
df10 = pd.read_csv('ODI 2024/final_players_ODI_2024_info.csv')
```

```python
print(df1)
print(df2)
print(df3)
print(df4)
print(df5)
print(df6)
print(df7)
print(df8)
print(df9)
print(df10)
```

```python
# Concatenate all DataFrames
all_players_df = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8,df9,df10])

# Count occurrences of each player
player_counts = all_players_df['Player Name'].value_counts()

# Display the result
print("Number of times each player appears across all DataFrames:")
player_counts
```

```python
# Convert player_counts to DataFrame
player_counts_df = player_counts.reset_index()
player_counts_df.columns = ['Player Name', 'Count']  # Rename columns for clarity

# Save to CSV file
player_counts_df.to_csv('player_appearance_counts.csv', index=False)

print("Player counts saved to 'player_appearance_counts.csv'")
player_counts_df
```

```python
filtered_players_df = player_counts_df[player_counts_df["Count"] >= 3]

# Merge filtered_players_df with all_players_df to get additional details
players_with_details = pd.merge(filtered_players_df, all_players_df, on="Player Name", how="left")
players_with_details = players_with_details.drop_duplicates(subset="Player Name")

# Display the result
players_with_details
```

```python
import pandas as pd
df1 = pd.read_csv('ipl 2020/combined_player_data.csv')
df2 = pd.read_csv('ipl 2021/combined_player_data.csv')
df3 = pd.read_csv('ipl 2022/combined_player_data.csv')
df4 = pd.read_csv('ipl 2023/combined_player_data.csv')
df5 = pd.read_csv('ipl 2024/combined_player_data.csv')
df6 = pd.read_csv('t20 2022/combined_player_data.csv')
df7 = pd.read_csv('t20 2024/combined_player_data.csv')
df8 = pd.read_csv('t20 2020/combined_player_data.csv')
df9 = pd.read_csv('ODI 2019/combined_player_data.csv')
df10 = pd.read_csv('ODI 2024/combined_player_data.csv')
df10 = pd.read_csv('ODI 2024/combined_player_data.csv')
players_with_details
```

```python
all_data = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8, df9, df10], ignore_index=True)

# Filter rows for the specified players
filtered_data = all_data[all_data['Player Name'].isin(players_with_details['Player Name'])]
filtered_data = filtered_data.sort_values(by='Player Name', ascending=True)
# Display the filtered dataframe
batsman_table = filtered_data[
    ['batsmanName', 'inning_batted', 'Average_SR', 'Average_Boundary_Percentage', 'total_runs', 'batting_average']
].dropna(subset=['batsmanName'])  # Drop rows where batsmanName is NaN
specific_players = [
    "Glenn Maxwell", "Jos Buttler", "David Warner",
    "Suryakumar Yadav", "Heinrich Klaasen", "Glenn Phillips", "Virat Kohli",
    "Marcus Stoinis", "Travis Head"
]
# Filter the batsman_table for the specific players
filtered_batsman_table = batsman_table[batsman_table['batsmanName'].isin(specific_players)]


average_stats = filtered_batsman_table.groupby('batsmanName').mean(numeric_only=True).reset_index()

# Display the resulting averages table
print("Average Stats for batsman:")
average_stats

```

```python
# Table for bowler details
bowler_table = filtered_data[
    ['bowlerName', 'inningbowled', 'avg_bowling_economy', 'avg_bowling_strike', 'average_bowling_avg', 'average_dot_ball_%']
].dropna(subset=['bowlerName'])  # Drop rows where bowlerName is NaN
bowler_table
average_stats = bowler_table.groupby('bowlerName').mean(numeric_only=True).reset_index()

# Display the resulting averages table
print("Average Stats for bowlers:")
average_stats
```
